```python
#checking for columns with NaN values
for col in anime.columns:
    missing_count = anime[col].isnull().sum()
    pct_missing = (anime[col].isnull().mean()) * 100
    dtype = anime[col].dtype
    print(f'{col} |  {dtype} | Missing: {missing_count} ({pct_missing:.2f}%)')
```

```python
#dropping columns with 60% or more nans
cols_to_drop = []
for col in anime.columns:
    missing_percent = anime[col].isnull().mean() * 100
    if missing_percent >= 60:
        cols_to_drop.append(col)
    else:
        pass  

# drop the selected columns
anime_cleaned = anime.drop(columns=cols_to_drop)
```
